```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
diabetes_data = pd.read_csv("diabetes.csv")
diabetes_data.head(7)
```

```python
diabetes_data.shape
```

```python
diabetes_data.info()
```

```python
diabetes_data.describe()
```

```python
x = diabetes_data.drop(['Outcome'], axis = 1)
y = diabetes_data["Outcome"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state =175)
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
print("Naive Bayes models' accuracy(in %) for test data is :", metrics.accuracy_score(y_test, y_pred)*100)

c_misclassified = (y_test != y_pred).sum()
print('Misclassified samples are : {}'.format(c_misclassified))
```

```python
y_pred_train = model.predict(X_train)
print("Naive Bayes model accuracy(in %) for train data is :", metrics.accuracy_score(y_train, y_pred_train)*100)

c_misclassified_t = (y_train != y_pred_train).sum()
print('Misclassified samples: {}'.format(c_misclassified_t))
```

```python
#THUS WE CAN SEE OUR MODEL TRAINING ACCURACY is to 76% and TESTING ACCURACY is to 75%
```
